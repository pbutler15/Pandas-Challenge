```python
#Gender Demographics
#Find the percentage and count of male, female, and other/non-disclosure players

# Group the file by Gender
gender_stats= HeroesOfPymoli_df.groupby("Gender")

# Calculate the total by gender
total_count_gender = gender_stats.nunique() ["SN"]

# To obtain the percentage use the total count by gender and divivde by the total number of players then multiply by 100
percentage_of_players = total_count_gender / player_total * 100

# Create data frame with all values
gender_demographics = pd.DataFrame({"Percentage of Players": percentage_of_players, "Total Count": total_count_gender})

# Format the values sorted by total count in descending order, and two decimal places for the percentage
gender_demographics.sort_values(["Total Count"], ascending = False).style.format({"Percentage of Players":"{:.2f}"})
```

```python
#Purchasing Analysis (Gender)

#Count the total number of purchases 
purchase_count = gender_stats["Purchase ID"].count()
purchase_count

#Calculate the average of the purchase count
average_purchase_price = gender_stats["Price"].mean()
average_purchase_price

#Calculate the total purchase value
total_purchase_value = gender_stats["Price"].sum()
total_purchase_value

#Calculate the total average purchase per person
per_person = total_purchase_value/total_count_gender
per_person

#Create dataframe and format accurate currency
purchase_analysis= pd.DataFrame({"Purchase Count": purchase_count, 
                                    "Average of Purchase Price": average_purchase_price,
                                    "Average of Purchase Value":total_purchase_value,
                                    "Average Purchase Per Person": per_person})

purchase_analysis["Average of Purchase Price"] = purchase_analysis["Average of Purchase Price"].map("${:,.2f}".format)
purchase_analysis["Average of Purchase Value"] = purchase_analysis["Average of Purchase Value"].map("${:,.2f}".format)
purchase_analysis["Average Purchase Per Person"] = purchase_analysis["Average Purchase Per Person"].map("${:,.2f}".format)
purchase_analysis

```

```python
#Age Demographics

#Establish bins for ages
age_bins= [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99]

age_labels= ["<10", "10 -14","15-19","20-24","25-29","30-34","35-39","40+"]

#Categorize the existing players using the age bins.
HeroesOfPymoli_df["Age Demographics"] = pd.cut(HeroesOfPymoli_df["Age"], bins = age_bins, labels = age_labels, include_lowest=True)
HeroesOfPymoli_df

#Add new data frame with "Age Demographics" and group it
age_group = HeroesOfPymoli_df.groupby("Age Demographics")
age_group

# Count total players by age category
total_age_count = age_group["SN"].nunique()
total_age_count

# Calculate percentages by age category 
percentage_by_age = round(total_age_count/player_total* 100,2).astype(str) + '%'
percentage_by_age

# Create data frame with obtained values
age_demographics = pd.DataFrame({"Percentage of Players": percentage_by_age, 
                                 "Total Count": total_age_count})
age_demographics

```

```python
#Purchasing Analysis (Age)

#Find the number of purchase Count
purchase_byage = age_group["Purchase ID"].count()
purchase_byage

#Average of Purchase price by age group
avg_purchase_byage = age_group["Price"].mean()
avg_purchase_byage

#Total Purchase value by age group
purchase_total= age_group["Price"].sum()
purchase_total

#Average of Purchase Count Per Person by age group
avg_perperson_age = purchase_total/total_age_count
avg_perperson_age

#Create dataframe for the values
age_demographics = pd.DataFrame({"Purchase Count": purchase_byage,
                                 "Average Purchase Price": avg_purchase_byage,
                                 "Total Purchase Value":purchase_total,
                                 "Average Purchase Total per Person": avg_perperson_age})
age_demographics

#Add currency and round values
age_demographics["Average Purchase Price"] = age_demographics["Average Purchase Price"].map("${:,.2f}".format)
age_demographics["Total Purchase Value"] = age_demographics["Total Purchase Value"].map("${:,.2f}".format)
age_demographics["Average Purchase Total per Person"] = age_demographics["Average Purchase Total per Person"].map("${:,.2f}".format)
age_demographics
```

```python
#Top Spenders

#Group spenders 
spend_group = HeroesOfPymoli_df.groupby("SN")
spend_group

#Calulate purchase count of spenders
spend_purchase_count = spend_group["Purchase ID"].count()
spend_purchase_count

#Calulate the avergae purchase price of spenders
avg_spenders_price = spend_group["Price"].mean()
avg_spenders_price

#Calculate total amount of top spenders
total_spender_amount = spend_group["Price"].sum()
total_spender_amount

#Create dataframe 
top_spenders = pd.DataFrame ({"Purchase Count":spend_purchase_count, "Average Purchase Price":avg_spenders_price,"Total Purchase Value":total_spender_amount})
top_spenders

# Sort in descending order to obtain top 5 spender names 
formatted_spenders = top_spenders.sort_values(["Total Purchase Value"], ascending=False).head()
formatted_spenders

#Add currency and round
formatted_spenders["Average Purchase Price"] = formatted_spenders["Average Purchase Price"].map("${:,.2f}".format)
formatted_spenders["Total Purchase Value"] = formatted_spenders["Total Purchase Value"].map("${:,.2f}".format)
formatted_spenders
```

```python
#Most Popular Items

#Retrieve item Id, item name and item price columns
item_list = HeroesOfPymoli_df [["Item ID", "Item Name", "Price"]]
item_list

#Groupby item ID and item name
items = HeroesOfPymoli_df.groupby(["Item ID","Item Name"])

#Count the number of times an item was purchased
number_ofpurchase = items["Price"].count()
number_ofpurchase

#Purchase value per item
purchase_value = (items["Price"].sum())
purchase_value

#Calculate total purchase value
item_price = purchase_value/number_ofpurchase
item_price

#Create dataframe
popular_items = pd.DataFrame({"Purchase Count":number_ofpurchase,"Item Price":item_price, "Total Purchase Value":purchase_value})
popular_items

# Sort in descending order to obtain top spender names and provide top 5 item names
popular_formatted = popular_items.sort_values(["Purchase Count"], ascending=False).head()
popular_formatted

#Set curency and round
popular_formatted["Item Price"] = popular_formatted["Item Price"].map("${:,.2f}".format)
popular_formatted["Total Purchase Value"] = popular_formatted["Total Purchase Value"].map("${:,.2f}".format)
popular_formatted
```

```python
#Most Profittable Items

#Use the dataframe above and srt to find the highest total purchase value
popular_formatted = popular_items.sort_values(["Total Purchase Value"],ascending=False).head()

#Set currency and round
popular_formatted["Item Price"] = popular_formatted["Item Price"].map("${:,.2f}".format)
popular_formatted["Total Purchase Value"] = popular_formatted["Total Purchase Value"].map("${:,.2f}".format)
popular_formatted
```
